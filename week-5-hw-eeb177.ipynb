```python
#Part3. Chapter 5. exercise. Percentage of amino acid residues, P1.



def percentage(protein, aminoacid):
    protein = protein.upper()
    aminoacid= aminoacid.upper()
    
    aacount= protein.count(aminoacid)
    aalength= len(protein)
    percentage= aacount * 100 / aalength
    return percentage


assert percentage("MSRSLLLRFLLFLLLLPPLP","M") == 5
assert percentage("MSRSLLLRFLLFLLLLPPLP","r") == 10
assert percentage("MSRSLLLRFLLFLLLLPPLP","L") == 50
assert percentage("MSRSLLLRFLLFLLLLPPLP","Y") == 0
```

```python
#Part3. Chapter 5. exercise. Percentage of amino acid residues, P2.

def aapercentage(protein, aalist=['A','I','L','M','F','W','Y','V']):
    protein=protein.upper()
    plength= len(protein)
    total =0
    
    for aa in aalist:
        aa=aa.upper()
        aacount=protein.count(aa)
        total += aacount
    
    percentage = total * 100 / plength
    return percentage

    
    
    
assert aapercentage("MSRSLLLRFLLFLLLLPPLP",["M"]) == 5
assert aapercentage("MSRSLLLRFLLFLLLLPPLP", ['M', 'L']) == 55
assert aapercentage("MSRSLLLRFLLFLLLLPPLP", ['F', 'S', 'L']) == 70
assert aapercentage("MSRSLLLRFLLFLLLLPPLP") == 65
```
